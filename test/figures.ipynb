# Figures
This notebook is used to create the figures that will be used in the game.

```python
import numpy as np

import matplotlib.pyplot as plt
```

## Coffee Brewing Chart 

```python
fig, ax = plt.subplots(figsize=(12,10))

ax.axvspan(18,22,color="orange",alpha=0.7)
ax.axhspan(1.15,1.35,color="orange",alpha=0.7)

ax.plot([14,20.5],[1.10,1.6],color="red",alpha=0.8)
ax.plot([14,23.5],[0.975,1.6],color="red",alpha=0.8)
ax.plot([14,26],[0.825,1.525],color="red",alpha=0.8)
ax.plot([14,26],[0.7,1.275],color="red",alpha=0.8)

ax.set_xlim([14,26])
ax.set_xticks(np.arange(14,27,1))
ax.set_ylim([0.8,1.6])

ax.grid(which="both")

plt.savefig("../images/figures/coffee_brewing.png")
plt.show()
plt.close()
```
